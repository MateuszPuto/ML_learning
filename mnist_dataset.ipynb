```python
import numpy as np
from tensorflow import keras
from keras.datasets import mnist
from keras import models
from keras import layers
import matplotlib.pyplot as plt
```

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
print("Shape of dataset: ", train_images.shape, test_images.shape)
print("Shape of labels: ", train_labels.shape, test_labels.shape)

plt.imshow(train_images[0], cmap = plt.cm.binary)
plt.show()

print("Example image: ", train_images[0])
print("Example label: ", train_labels[0])
```

```python
#data preprocessing

train_images = train_images.reshape(60000, 28, 28, 1)
test_images = test_images.reshape(10000, 28, 28, 1)

train_images = train_images.astype("float32") / 255
test_images = test_images.astype("float32") / 255

train_labels = keras.utils.to_categorical(train_labels)
test_labels = keras.utils.to_categorical(test_labels)
```

```python
print("Shape of dataset: ", train_images.shape, test_images.shape)
print("Shape of labels: ", train_labels.shape, test_labels.shape)

print("Example image: ", train_images[0])
print("Example label: ", train_labels[0])
```

```python
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(10, activation='softmax'))

model.build()
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['acc'])
```

```python
model.fit(train_images, train_labels, batch_size=128, epochs=5)
```

```python
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=0)
print("Loss: ", test_loss, ", accuracy: ", test_acc)
```

```python
predictions = model.predict(train_images)
print("First image predicted class: ", np.argmax(predictions[0]))
```
