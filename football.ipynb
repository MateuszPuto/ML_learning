```python
def odds_to_prob(y):
    return y / (y + 1)

def prob_to_odd(x):
    return x / (1 - x)

odd_home = ginf['odd_h']
odd_draw = ginf['odd_d']
odd_away = ginf['odd_a']

print('Example odds to probabilities:')
print(odd_home[0], '\t', 1-odds_to_prob(odd_home[0]), '\t\t\t', prob_to_odd(odds_to_prob(odd_home[0])))
print(odd_draw[0], '\t', 1-odds_to_prob(odd_draw[0]), '\t\t', prob_to_odd(odds_to_prob(odd_draw[0])))
print(odd_away[0], '\t', 1-odds_to_prob(odd_away[0]), '\t\t', prob_to_odd(odds_to_prob(odd_away[0])), end='\n\n')

values = []
for i in range(0, ginf.shape[0]):
    total_value = 1-odds_to_prob(odd_home[i]) + 1-odds_to_prob(odd_draw[i]) + 1-odds_to_prob(odd_away[i])
    values.append(total_value)

print(f'Average odds of winning against bookmaker (whole dataset): {100 * sum(values) / len(values):.2f}%')
```

```python
def to_outcome(proportion):
    out = None
    if proportion > 1.0:
        out = 1
    if proportion < 1.0:
        out = -1
    else:
        out = 0

    return out

ginf['fthg/ftag'] = ginf['fthg'] / ginf['ftag']
ginf['ftag/fthg'] = ginf['ftag'] / ginf['fthg']
ginf['winloss'] = [to_outcome(x) for x in ginf['fthg/ftag']]

goals_and_odds = ginf[['odd_h', 'odd_d', 'odd_a', 'fthg/ftag', 'ftag/fthg', 'winloss']]

corr = goals_and_odds.corr()
print(corr)

plt.imshow(corr)
plt.show()
```

- Wysoka korelacja między kursem bukmacherskim dla remisu a kursem dla drużyny gości (0.86).

- Proporcja bramek, które miały miejsce w meczu, jest odwrotnie skorelowana z kursem bukmacherskim.

- Poziom korelacji między tym czego spodziewali się bukmacherzy przed meczem a wynikiem meczu wskazuje na średni lub słaby poziom możliwości przewidzenia wyniku meczu przed jego rozegraniem.

```python
x.reset_index(drop=True, inplace=True)
y.reset_index(drop=True, inplace=True)
z.reset_index(drop=True, inplace=True)
```

```python
from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)
```

```python
tree.plot_tree(clf, max_depth=2, fontsize=8)
```

```python
clf.predict(x.loc[0:100])
```

```python
location = pd.get_dummies(x['location'])
bodypart = pd.get_dummies(x['bodypart'])
assist_method = pd.get_dummies(x['assist_method'])
situation = pd.get_dummies(x['situation'])
```

```python
dummies = [location, bodypart, assist_method, situation]
one_hot = pd.concat(dummies, axis=1, ignore_index=True)

one_hot
```

```python
shot_place = pd.get_dummies(z['shot_place'])
shot_outcome = pd.get_dummies(z['shot_outcome'])
```

```python
dummies = [shot_place, shot_outcome]
dependant = pd.concat(dummies, axis=1, ignore_index=True)

dependant
```

```python
validation_split_size = int(one_hot.shape[0]/10)
print(validation_split_size)

clf2 = tree.DecisionTreeClassifier()
clf2 = clf.fit(one_hot[validation_split_size:], y[validation_split_size:])
```

```python
tree.plot_tree(clf2, max_depth=2, fontsize=8)
```

```python
clf2.predict(one_hot.loc[:100])
```

```python
clf2.score(one_hot.loc[:validation_split_size], y.loc[:validation_split_size])
```

```python
clf3 = tree.DecisionTreeClassifier()
clf3 = clf.fit(dependant[validation_split_size:], y[validation_split_size:])
```

```python
clf3.score(dependant.loc[:validation_split_size], y.loc[:validation_split_size])
```

```python
dummies = [one_hot, dependant]
full_information = pd.concat(dummies, axis=1, ignore_index=True)

full_information
```

```python
clf4 = tree.DecisionTreeClassifier()
clf4 = clf.fit(full_information[validation_split_size:], y[validation_split_size:])
```

```python
clf4.score(full_information.loc[:validation_split_size], y.loc[:validation_split_size])
```

```python
data_where_goal = full_information[y['is_goal'] == 1]

clf4.score(data_where_goal, y[y['is_goal'] == 1])
```

```python
dummies = [one_hot, x['time']]
with_match_time = pd.concat(dummies, axis=1, ignore_index=True)

with_match_time
```

```python
clf5 = tree.DecisionTreeClassifier()
clf5 = clf.fit(with_match_time[validation_split_size:], y[validation_split_size:])
```

```python
clf5.score(with_match_time.loc[:validation_split_size], y.loc[:validation_split_size])
```

The 4'th classifier using full information without game time works best on this data. It classifies 94% of shots correctly and is able to label 64,6% of real goals as scoring, even though only obout 2% of datapoints have positive label. The downgrade in quality of Tree Classifier when game time data is used may be attributed to injecting low quality information, which does not affect the probability of goal, into training set.
