```python
def odds_to_prob(y):
    return y / (y + 1)

def prob_to_odd(x):
    return x / (1 - x)

odd_home = ginf['odd_h']
odd_draw = ginf['odd_d']
odd_away = ginf['odd_a']

print('Example odds to probabilities:\n')
print('Outcome\t\t', 'Odds\t', 'Implied probability\t\t', 'Check (odds->prob.->odds)')
print('_______________________________________________________________________')
print('Home win\t', odd_home[0], '\t', odds_to_prob(odd_home[0]), '\t\t\t', prob_to_odd(odds_to_prob(odd_home[0])))
print('Draw\t\t', odd_draw[0], '\t', odds_to_prob(odd_draw[0]), '\t\t', prob_to_odd(odds_to_prob(odd_draw[0])))
print('Away win\t', odd_away[0], '\t', odds_to_prob(odd_away[0]), '\t\t', prob_to_odd(odds_to_prob(odd_away[0])), end='\n\n')

avg_odd = []
for i in range(0, ginf.shape[0]):
    avg_odd.append(1/3 * (1+odd_home[i]) + 1/3 * (1+odd_draw[i]) + 1/3 * (1+odd_away[i]))

print(f'Zakładając, że obstawiamy wszystkie możliwe rezultaty poprzez równy podział $1.00 na trzy części, nasza średnia wygrana wyniosłaby ${odds_to_prob(sum(avg_odd) / len(avg_odd)):.2f}.')
```

```python
def to_outcome(proportion):
    out = None
    if proportion > 1.0:
        out = 1
    if proportion < 1.0:
        out = -1
    else:
        out = 0

    return out

ginf['fthg/ftag'] = ginf['fthg'] / ginf['ftag']
ginf['ftag/fthg'] = ginf['ftag'] / ginf['fthg']
ginf['winloss'] = [to_outcome(x) for x in ginf['fthg/ftag']]

goals_and_odds = ginf[['odd_h', 'odd_d', 'odd_a', 'fthg/ftag', 'ftag/fthg', 'winloss']]

corr = goals_and_odds.corr()
print(corr)

plt.imshow(corr)
plt.show()
```

- Wysoka korelacja między kursem bukmacherskim dla remisu a kursem dla drużyny gości (0.86).

- Proporcja bramek, które miały miejsce w meczu, jest odwrotnie skorelowana z kursem bukmacherskim.

- Poziom korelacji między tym czego spodziewali się bukmacherzy przed meczem a wynikiem meczu wskazuje na średni lub słaby poziom możliwości przewidzenia wyniku meczu przed jego rozegraniem.

# Wykorzystanie zmiennych kategorycznych

```python
location = pd.get_dummies(x['location'])
bodypart = pd.get_dummies(x['bodypart'])
assist_method = pd.get_dummies(x['assist_method'])
situation = pd.get_dummies(x['situation'])
```

```python
dummies = [location, bodypart, assist_method, situation]
one_hot = pd.concat(dummies, axis=1, ignore_index=True)

one_hot
```

```python
shot_place = pd.get_dummies(z['shot_place'])
shot_outcome = pd.get_dummies(z['shot_outcome'])
```

```python
dummies = [shot_place, shot_outcome]
dependant = pd.concat(dummies, axis=1, ignore_index=True)

dependant
```

Clf2 - podział na zbiór treningowy i testowy, działa na zmiennych kategorycznych

```python
validation_split_size = int(one_hot.shape[0]/10)
print(validation_split_size)

clf2 = tree.DecisionTreeClassifier()
clf2 = clf.fit(one_hot[validation_split_size:], y[validation_split_size:])
```

```python
tree.plot_tree(clf2, max_depth=2, fontsize=8)
```

```python
clf2.predict(one_hot.loc[:100])
```

```python
true_class = clf2.score(one_hot.loc[:validation_split_size], y.loc[:validation_split_size])
```

```python
data_where_goal = one_hot[y['is_goal'] == 1]

real_goal_correct_label = clf2.score(data_where_goal, y[y['is_goal'] == 1])
```

```python
print(f'Klasyfikator podaje poprawną klasę w {100 * true_class:.2f}% przypadków.')
print(f'Klasyfikator przewiduje gola w {100 * real_goal_correct_label:.2f}% przypadków.')
```

```python
dummies = [one_hot, dependant]
full_information = pd.concat(dummies, axis=1, ignore_index=True)

full_information
```

Clf3 - błędne założenie modelu, wykorzystujemy zmienne zależne

```python
clf3 = tree.DecisionTreeClassifier()
clf3 = clf.fit(full_information[validation_split_size:], y[validation_split_size:])
```

```python
true_class = clf3.score(full_information.loc[:validation_split_size], y.loc[:validation_split_size])
```

```python
data_where_goal = full_information[y['is_goal'] == 1]

real_goal_correct_label = clf3.score(data_where_goal, y[y['is_goal'] == 1])
```

```python
print(f'Klasyfikator podaje poprawną klasę w {100 * true_class:.2f}% przypadków.')
print(f'Klasyfikator przewiduje gola w {100 * real_goal_correct_label:.2f}% przypadków.')
```

Przy wykorzystaniu zmiennych zależnych osiągany wynik znacząco się poprawił, ale wykorzystujemy tu informacje, których nie możemy znać przed wykonaniem strzału.

```python
dummies = [one_hot, x['time']]
with_match_time = pd.concat(dummies, axis=1, ignore_index=True)

with_match_time
```

Clf4 - klasyfikator tak jak 'clf2' wykorzystujący dodatkowo czas meczu

```python
clf4 = tree.DecisionTreeClassifier()
clf4 = clf.fit(with_match_time[validation_split_size:], y[validation_split_size:])
```

```python
true_class = clf4.score(with_match_time.loc[:validation_split_size], y.loc[:validation_split_size])
```

```python
data_where_goal = with_match_time[y['is_goal'] == 1]

real_goal_correct_label = clf4.score(data_where_goal, y[y['is_goal'] == 1])
```

```python
print(f'Klasyfikator podaje poprawną klasę w {100 * true_class:.2f}% przypadków.')
print(f'Klasyfikator przewiduje gola w {100 * real_goal_correct_label:.2f}% przypadków.')
```

Klasyfikator 'Clf4' osiąga najlepsze wyniki przewidywania wyniku strzału na bramkę. Model podaje poprawną binarną klasę wyniku w około 90% przypadków, co jest niższym wynikiem od 'Clf2'. Zaletą tego klasyfikatora jest jednak większa sprawność dla strzałów, które naprawdę okazały się golami. W tym przypadku osiąga on poprawę o około 5 punktów procentowych. Jest to ważna metryka ze względu na fakt, iż tylko w wyniku około 2,5% strzałów na bramkę pada gol.
