```python
def odds_to_prob(y):
    return y / (y + 1)

def prob_to_odd(x):
    return x / (1 - x)

odd_home = ginf['odd_h']
odd_draw = ginf['odd_d']
odd_away = ginf['odd_a']

print('Example odds to probabilities:\n')
print('Outcome\t\t', 'Odds\t', 'Implied probability\t\t', 'Check (odds->prob.->odds)')
print('_______________________________________________________________________')
print('Home win\t', odd_home[0], '\t', odds_to_prob(odd_home[0]), '\t\t\t', prob_to_odd(odds_to_prob(odd_home[0])))
print('Draw\t\t', odd_draw[0], '\t', odds_to_prob(odd_draw[0]), '\t\t', prob_to_odd(odds_to_prob(odd_draw[0])))
print('Away win\t', odd_away[0], '\t', odds_to_prob(odd_away[0]), '\t\t', prob_to_odd(odds_to_prob(odd_away[0])), end='\n\n')

avg_odd = []
for i in range(0, ginf.shape[0]):
    avg_odd.append(1/3 * (1+odd_home[i]) + 1/3 * (1+odd_draw[i]) + 1/3 * (1+odd_away[i]))

print(f'ZakÅ‚adajÄ…c, Å¼e obstawiamy wszystkie moÅ¼liwe rezultaty poprzez rÃ³wny podziaÅ‚ $1.00 na trzy czÄ™Å›ci, nasza Å›rednia wygrana wyniosÅ‚aby ${odds_to_prob(sum(avg_odd) / len(avg_odd)):.2f}.')
```

```python
def to_outcome(proportion):
    out = None
    if proportion > 1.0:
        out = 1
    if proportion < 1.0:
        out = -1
    else:
        out = 0

    return out

ginf['fthg/ftag'] = ginf['fthg'] / ginf['ftag']
ginf['ftag/fthg'] = ginf['ftag'] / ginf['fthg']
ginf['winloss'] = [to_outcome(x) for x in ginf['fthg/ftag']]

goals_and_odds = ginf[['odd_h', 'odd_d', 'odd_a', 'fthg/ftag', 'ftag/fthg', 'winloss']]

corr = goals_and_odds.corr()
print(corr)

plt.imshow(corr)
plt.show()
```

- Wysoka korelacja miÄ™dzy kursem bukmacherskim dla remisu a kursem dla druÅ¼yny goÅ›ci (0.86).

- Proporcja bramek, ktÃ³re miaÅ‚y miejsce w meczu, jest odwrotnie skorelowana z kursem bukmacherskim.

- Poziom korelacji miÄ™dzy tym czego spodziewali siÄ™ bukmacherzy przed meczem a wynikiem meczu wskazuje na Å›redni lub sÅ‚aby poziom moÅ¼liwoÅ›ci przewidzenia wyniku meczu przed jego rozegraniem.

# Wykorzystanie zmiennych kategorycznych

```python
location = pd.get_dummies(x['location'])
bodypart = pd.get_dummies(x['bodypart'])
assist_method = pd.get_dummies(x['assist_method'])
situation = pd.get_dummies(x['situation'])
```

```python
dummies = [location, bodypart, assist_method, situation]
one_hot = pd.concat(dummies, axis=1, ignore_index=True)

one_hot
```

```python
shot_place = pd.get_dummies(z['shot_place'])
shot_outcome = pd.get_dummies(z['shot_outcome'])
```

```python
dummies = [shot_place, shot_outcome]
dependant = pd.concat(dummies, axis=1, ignore_index=True)

dependant
```

Clf2 - podziaÅ‚Â na zbiÃ³r treningowy i testowy, dziaÅ‚a na zmiennych kategorycznych

```python
validation_split_size = int(one_hot.shape[0]/10)
print(validation_split_size)

clf2 = tree.DecisionTreeClassifier()
clf2 = clf.fit(one_hot[validation_split_size:], y[validation_split_size:])
```

```python
tree.plot_tree(clf2, max_depth=2, fontsize=8)
```

```python
clf2.predict(one_hot.loc[:100])
```

```python
true_class = clf2.score(one_hot.loc[:validation_split_size], y.loc[:validation_split_size])
```

```python
data_where_goal = one_hot[y['is_goal'] == 1]

real_goal_correct_label = clf2.score(data_where_goal, y[y['is_goal'] == 1])
```

```python
print(f'Klasyfikator podaje poprawnÄ… klasÄ™Â w {100 * true_class:.2f}% przypadkÃ³w.')
print(f'Klasyfikator przewiduje gola w {100 * real_goal_correct_label:.2f}% przypadkÃ³w.')
```

```python
dummies = [one_hot, dependant]
full_information = pd.concat(dummies, axis=1, ignore_index=True)

full_information
```

Przy wykorzystaniu zmiennych zaleÅ¼nych osiÄ…gany wynik znaczÄ…co siÄ™Â poprawiÅ‚, ale wykorzystujemy tu informacje, ktÃ³rych nie moÅ¼emy znaÄ‡ przed wykonaniem strzaÅ‚u.
